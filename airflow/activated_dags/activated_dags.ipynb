```python
AIRFLOW_API_URL = 'http://localhost:8080/api/v1'
AIRFLOW_USERNAME = "admin"
AIRFLOW_PASSWORD = "admin"

# Basic Authentication (username/password)
auth = (AIRFLOW_USERNAME, AIRFLOW_PASSWORD)
for dag in dagname:
    # Endpoint to activate a DAG
    endpoint_activate = f'{AIRFLOW_API_URL}/dags/{dag}'

    activate_payload = {'is_paused': False}

    # Send a PATCH request to activate the DAG
    response = requests.patch(endpoint_activate, json=activate_payload, auth=auth)
    print(response.text)
```

```python
import requests

# Configuration
AIRFLOW_API_URL = 'http://localhost:8080/api/v1'
AIRFLOW_DAG_NAME = "train_modele_prediction"
AIRFLOW_USERNAME = "admin"
AIRFLOW_PASSWORD = "admin"

# Basic Authentication (username/password)
auth = (AIRFLOW_USERNAME, AIRFLOW_PASSWORD)

# Correct Endpoint to activate a DAG
endpoint_activate = f'{AIRFLOW_API_URL}/dags/{AIRFLOW_DAG_NAME}'

# Payload to unpause the DAG
activate_payload = {'is_paused': False}

# Send a PATCH request to activate the DAG
response = requests.patch(endpoint_activate, json=activate_payload, auth=auth)
if response.status_code == 200:
    print("DAG activated successfully.")
else:
    print(f"Failed to activate DAG: {response.status_code} {response.text}")
```
